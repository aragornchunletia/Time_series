```python
def change_name (name):
    if 'TG' in name:
        name = name.split(' ')
        return 'TG_' + str(name[1])
    return 'ril_no_'+str(name)
```

```python
df_khw.drop(['wt' , 'HKW'] , axis = 1 , inplace= True)
df_khw.head()
```

```python

df_s24.columns = df_s24.iloc[0]
df_s24.drop(index=0 , axis= 0 , inplace= True)
df_s24.head()
```

```python
df_s24['RIL'] = df_s24['RIL'].apply(lambda x : change_name(x))
df_s24.fillna(0.0 , inplace= True)
df_s24
```

```python
days_s24 = ['day0', 'day3', 'day6', 'day9', 'day12', 'day15',
       'day18', 'day21', 'day24', 'day27', 'day30', 'day36', 'day39', 'day42',
       'day45', 'day48']
```

```python
days_kh23 = [ 'day0', 'day3', 'day6', 'day9', 'day12', 'day15',
       'day18', 'day21', 'day24', 'day27']
```

```python
def to_pct(df , days):
    for day in days:
        df[day] = round((df[day]/df['no_seeds'])*100 , 2)
    return df
```

```python
df_khw = to_pct(df_khw , days_kh23)
df_s24 = to_pct(df_s24 , days_s24)
```

```python
df_khw.drop('no_seeds' , axis = 1 , inplace=True)
df_khw.head()
```

```python
df_s24.fillna(0.0 , inplace= True)
df_s24.drop('no_seeds' , axis = 1 , inplace= True)
df_s24.head()
```
