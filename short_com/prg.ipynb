```python
def create_csv(row):
    folder = r'C:\Users\mchun\CN-ML\time_series\progress'
    name = "rill_no_" + row.iloc[0] if row.iloc[0].isdigit() else row.iloc[0]
    sowing_date = row.iloc[1]
    harvesting_date = row.iloc[2]
    total_seeds = row.iloc[3]
    doGermin = []
    noSeed = []
    for i in range(4 , len(row)):
        if i%2 == 1:
            noSeed.append(row.iloc[i])
        else:
            doGermin.append(row.iloc[i])

    df = pd.DataFrame({
        "name": [name]*len(noSeed),
        "sowing_date": [sowing_date]*len(noSeed),
        "harvesting_date":[harvesting_date]*len(noSeed),
        "total_seeds":[total_seeds]*len(noSeed),
        "date_of_germination" : doGermin,
        "number_of_seeds": noSeed,
    })

    df.Name = name
    csv_file = name+".csv"
    path = os.path.join(folder,csv_file)
    df.to_csv(path,sep=",",index=False)
    print(f"Saved to path->{path}!")

```

```python
df.apply(create_csv, axis=1)
```
